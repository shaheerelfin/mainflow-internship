```python
import pandas as pd
```

```python
data = pd.read_csv("student-mat.csv", sep=";")
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
data.dtypes
```

```python
data.shape
```

```python
newdata=data.dropna
```

```python
data=data.drop_duplicates(inplace=True)
```

```python
average = data['G3'].mean()
print(f"The average score in math (G3): {average}")
```

```python
students_above_15 = (data['G3'] > 15).sum()
print(f"Number of students who scored above 15 in G3: {students_above_15}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
studytime_g3_corr = data[['studytime', 'G3']].corr().iloc[0, 1]
print(f"Correlation between study time and final grade (G3): {studytime_g3_corr}")
```

```python
# 4. Which gender has a higher average final grade (G3)?
average_g3_by_gender = data.groupby('sex')['G3'].mean()

# Display the result
print("Average G3 score by gender:")
print(average_g3_by_gender)

# Determine which gender has the higher average
higher_gender = average_g3_by_gender.idxmax()
print(f"The gender with the higher average final grade (G3) is: {higher_gender}")
```

```python
# 5. Plot a histogram of final grades (G3)
plt.figure(figsize=(8, 6))
plt.hist(data['G3'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Final Grades (G3)', fontsize=14)
plt.xlabel('Final Grade (G3)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 6. Create a scatter plot between study time and final grade (G3)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='studytime', y='G3', data=data, color='blue')
plt.title('Scatter Plot: StudyTime vs Final Grade', fontsize=14)
plt.xlabel('StudyTime ', fontsize=12)
plt.ylabel('Final Grade (G3)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# 7. Create a bar chart comparing the average scores of male and female students
plt.figure(figsize=(8, 6))
average_g3_by_gender.plot(kind='bar', color=['blue', 'pink'], edgecolor='black', alpha=0.7)
plt.title('Average Final Grades by Gender', fontsize=14)
plt.ylabel('Average Final Grade (G3)', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
